## Ход работы

#### 0. Импортируем необходимые библиотеки

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### 1. Откроем датасет через пандас

```python
table = pd.read_csv('sales_data.csv')
```

#### 2.2. Покажем последние 5 элементов дата сета

```python
table.info()
```

#### 4. Укажем существующую метрику в дата сете

```python
table['Revenue']
```

#### 5. Для поля Order Quantity найдём количество элементов, среднее значение, минимум, максимум, стандартное отклонение и медиану 

```python
# 50% - медиана
table['Order_Quantity'].describe(percentiles = [0.5])
```

#### 6. Построим гистограмму для поля Order Quantity

```python
plt.figure()
sns.histplot(data = table, x = 'Order_Quantity', discrete = True, shrink = 0.8, edgecolor = None, zorder = 2)
plt.grid(axis = 'y', alpha = 0.3, zorder = 1)
plt.xticks(range(0, 33, 2))
plt.show()
```

#### 7. Посчитаем количество продуктов каждого типа и построим круговую диаграмму

```python
product_category = table.groupby('Product_Category').count().iloc[:, 0]
labels = product_category.index.tolist()
values = product_category.tolist()
labels_values = []
for index in range(3):
    labels_values += [f'{labels[index]} ({values[index]})']
```

```python
plt.figure()
plt.pie(product_category, labels = labels, autopct='%.0f%%')
plt.legend(labels_values, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

#### 8. Выведем все локации, где происходила продажа (страна и штаты)

```python
locations = table.groupby('Country')
output = locations['State'].unique().apply(', '.join)
df_output = output.reset_index()
df_output.columns = ['Country', 'States']
with pd.option_context('display.max_colwidth', None):
    display(df_output)
```

#### 9. Выведем все продукты компании

```python
products = pd.DataFrame(table['Product'].unique())
with pd.option_context('display.max_rows', None):
    display(products)
```

###### Продукты по подкатегориям

```python
products = table.groupby('Sub_Category')
all_products = products['Product'].unique().apply(', '.join).reset_index()
all_products.columns = ['Sub_Category', 'Product']
with pd.option_context('display.max_colwidth', None):
    display(all_products)
```

#### 10. Выведем столбчатую диаграмму 10 самых продаваемых продуктов

```python
most_ordering_series = table.groupby('Product')['Order_Quantity'].sum().nlargest(10)
most_ordering = most_ordering_series.reset_index()
most_ordering.columns = ['Product', 'Quantity']

plt.figure()
plt.grid(axis='y', alpha=0.3, zorder=1)
sns.barplot(data = most_ordering, x = 'Product', y = 'Quantity', zorder=2)
plt.xticks(rotation=45, ha='right')
plt.show()
```
